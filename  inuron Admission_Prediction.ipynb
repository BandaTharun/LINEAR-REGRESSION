## Lasso linear regression

```python
#lasso regression this helps the model to predict better with test data 
# so this will reduse the m cofficent and add alpha or landa to it 
#the formula is l=(y_org-y_pred)+lamda(|M|)
# to find the m value for lasso m=sum((y_org-y_pred)(y_org-Yorg_mean)-lamda)/sum(x_org-x_mean)**2
```

```python
#inshelly alpha is None beacouse we are not giving alpha value ,
#we are telling the model to  find it
# heare cv means cross validiation this will divide the given the data into the 
# menshiond numbers  and this will train and test the model ,frist it will take  
# frist four data to train and  last one for test and
# next it will take last four for train
# and frist one for test ,this will do the number of iter menshiond 
#in inmax_iter

# by doing more iterations the model will find the best alpha or lamda value


## the job of lassocv is to find the best alpha value in given multiple values

lassocv=LassoCV(alphas=None,cv=5,max_iter=2000000,normalize=True)
lassocv.fit(x_train,y_train)
```
